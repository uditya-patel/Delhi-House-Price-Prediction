```python
Q1 = new_df.quantile(0.25)
#print(Q1)
Q3 = new_df.quantile(0.75)
#print(Q3)
IQR = Q3 - Q1
print(IQR)
#print(Q1 - (1.5 * IQR))
#print(Q3 + (1.5 * IQR))
sns.boxplot(new_df['Per_Sqft'])
new_df.shape
```

```python
#REMOVES OUTLIERS PRESENT IN THE DATAFRAME USING IQR METHOD *************
new_df = new_df[~((new_df < (Q1 - 1.5 * IQR)) |(new_df > (Q3 + 1.5 * IQR))).any(axis=1)]
new_df.shape
```

```python
#'X' is the independent variable or feature.
#'y' is the dependent variable or feature.
X = new_df[['Area', 'BHK', 'Bathroom', 'Parking', 'Per_Sqft']].values
y = new_df[['Price']].values
```

***Splitting the dataset in training and test set***

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train)
print(x_train.shape)
```

```python
print(x_test)
print(x_test.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
print(y_test)
print(y_test.shape)
```

***Feature Scaling***

```python
#on performing feature scaling on training and testing set, both sets will be on the same scale.
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

### MULTIPLE LINEAR REGRESSION###

```python
#Training the model
from sklearn.linear_model import LinearRegression
lregressor = LinearRegression()
lregressor.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score, confusion_matrix
```

```python
#Train set result score.
y_pred1_train = sc_y.inverse_transform(lregressor.predict(x_train))
r1_train = r2_score(sc_y.inverse_transform(y_train), y_pred1_train)
print(r1_train)
```

```python
y_pred1_test = sc_y.inverse_transform(lregressor.predict(x_test))
```

```python
#Test set result score.
r1_test = r2_score(sc_y.inverse_transform(y_test), y_pred1_test)
print(r1_test)
```

### POLYNOMIAL REGRESSION###

```python
#Training the model
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(3)
x_poly = poly_reg.fit_transform(x_train)
lregressor.fit(x_poly, y_train)
```

```python
#Train set result score
y_pred2_train = sc_y.inverse_transform(lregressor.predict(poly_reg.transform(x_train)))
r2_train = r2_score(sc_y.inverse_transform(y_train), y_pred2_train)
print(r2_train)
```

```python
#Test set result score
y_pred2_test = sc_y.inverse_transform(lregressor.predict(poly_reg.transform(x_test)))
r2_test = r2_score(sc_y.inverse_transform(y_test), y_pred2_test)
print(r2_test)
```

```python
#np.set_printoptions(precision=2)
#print(np.concatenate((y_pred2_test.reshape(len(y_pred2_test),1), y_test.reshape(len(y_test),1)),1))
```

### DECISION TREE REGRESSOR###

```python
#Traing the model.
from sklearn.tree import DecisionTreeRegressor
DTregressor = DecisionTreeRegressor(random_state=0)
DTregressor.fit(x_train, y_train)
```

```python
#Train set result score
y_pred3_train = sc_y.inverse_transform(DTregressor.predict(x_train))
r3_train = r2_score(sc_y.inverse_transform(y_train), y_pred3_train)
print(r3_train)
```

```python
#Test set result score
y_pred3_test = sc_y.inverse_transform(DTregressor.predict(x_test))
r3_test = r2_score(sc_y.inverse_transform(y_test), y_pred3_test)
print(r3_test)
```

### RANDOM FOREST REGRESSOR###

```python
#Training the model.
from sklearn.ensemble import RandomForestRegressor
RFregressor = RandomForestRegressor(n_estimators=10, random_state=0)
RFregressor.fit(x_train, y_train)
```

```python
#Train set result score
y_pred4_train = sc_y.inverse_transform(RFregressor.predict(x_train))
r4_train = r2_score(sc_y.inverse_transform(y_train), y_pred4_train)
print(r4_train)
```

```python
#Test set result score
y_pred4_test = sc_y.inverse_transform(RFregressor.predict(x_test))
r4_test = r2_score(sc_y.inverse_transform(y_test), y_pred4_test)
print(r4_test)
```
